```python
import pandas as pd
import numpy as np
```

```python
# X: Features
X_train= pd.read_csv('X_train.csv')
X_test= pd.read_csv('X_test.csv')

# y: Labels for classification
y_train = pd.read_csv('y_train.csv', skiprows=1, names=['Diagnosis'])
y_test= pd.read_csv('y_test.csv', skiprows=1, names=['Diagnosis'])

# y reg: Numerical values for regression
y_train_reg= pd.read_csv('y_train_regression.csv', skiprows=1, names=['T_Score'])
y_test_reg= pd.read_csv('y_test_regression.csv', skiprows=1, names=['T_Score'])
```

```python
train_df = pd.concat([X_train, y_train, y_train_reg], axis=1) # Concatenate horizontally
test_df = pd.concat([X_test, y_test, y_test_reg], axis=1)
df = pd.concat([train_df, test_df], axis=0).reset_index(drop=True)# Concatenate vertically

df
```

```python
# df = pd.read_csv('your_data.csv')  # Si necesitas cargar los datos desde un archivo CSV

# Separar los datos por género
df_male = df[df['RIAGENDR'] == 1]
df_female = df[df['RIAGENDR'] == 2]

# Definir las columnas de interés
columns_of_interest = ['RIDAGEYR', 'BMXWT', 'PAXMINSB', 'PAXMINLPA', 'PAXMINMPA', 'PAXMINVPA', 'PAXMINVVPA', 'T_Score']

# Crear un DataFrame vacío para almacenar los resultados
summary_stats = pd.DataFrame(columns=['Variable', 'Sexo', 'count', 'Media', 'Std', 'Min', '25%', '50%', '75%', 'Max'])

# Función para calcular los estadísticos descriptivos
def calculate_summary_stats(df, gender):
    stats = df[columns_of_interest].describe().transpose()
    stats['Variable'] = stats.index
    stats['Sexo'] = gender
    return stats.reset_index(drop=True)

# Calcular los estadísticos descriptivos para hombres y mujeres
stats_male = calculate_summary_stats(df_male, 'Hombres')
stats_female = calculate_summary_stats(df_female, 'Mujeres')

# Concatenar los resultados
summary_stats = pd.concat([stats_male, stats_female], axis=0).reset_index(drop=True)

# Seleccionar y renombrar las columnas
summary_stats = summary_stats[['Variable', 'Sexo', 'count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']]
summary_stats.columns = ['Variable', 'Sexo', 'count', 'Mean', 'Std', 'Min', '25%', '50%', '75%', 'Max']


```

```python
summary_stats.to_csv('summary_stats.csv', index=False) # Guardar los resultados en un archivo CSV
```
